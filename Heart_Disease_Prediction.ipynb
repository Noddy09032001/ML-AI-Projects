According to resources

Stable Angina: Typically matches **"0: Typical angina"** because it involves chest pain related to decreased blood supply to the heart, often due to exertion.

NSTE-ACS (Non-ST Elevation Acute Coronary Syndrome): This aligns with **"1: Atypical angina"**, as it refers to atypical chest pain that may be caused by factors not directly associated with heart disease, though it may suggest underlying risk.

Noncardiac: This would correspond to **"2: Non-anginal pain"**, which is chest pain due to non-heart-related issues, such as esophageal spasms.

STEMI (ST-Elevation Myocardial Infarction): This is an acute heart attack condition but isn’t directly represented in the provided classification (0–3). However, **"3: Asymptomatic"** may be added here if someone is at risk but shows no symptoms.

```python
# splitting the dataset for generation of the model

x = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
display(x)
display(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

models = {"Logistic Regression": LogisticRegression(),
         "KNN": KNeighborsClassifier(),
         "Random Forest": RandomForestClassifier()}

def generate_model(models, X_train, X_test, y_train, y_test):
  np.random.seed(42)
  model_scores = {}
  for name, model in models.items():
    model.fit(X_train, y_train)
    model_scores[name]=model.score(X_test, y_test)
  return model_scores

# calling the function to get the scores of the models

model_scores = generate_model(models=models, X_train=X_train, X_test=X_test, y_train=y_train, y_test=y_test)
model_scores
```

```python
def predictions_generation(models, X_test, y):
  for name, model in models.items():
    y_pred = model.predict(X_test)
    y_pred_df = pd.DataFrame(y_pred, index=y_test.index, columns=["Predictions"])
    prediction_data = pd.concat([X_test,y_pred_df, y_test.rename("Actual")], axis=1)# combining the original and predicted data
    print(f"\n\nPredictions for the model: {name}")
    display(prediction_data)
    print(f"Predictions for the model {name}: {y_pred}")

predictions_generation(models = models, X_test=X_test, y = y)
```

```python
# comparing all the models
model_compare = pd.DataFrame(model_scores, index=["accuracy"])
model_compare.T.plot.bar()
```
